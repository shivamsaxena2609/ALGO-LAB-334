# 1. Growth Of Function
Time Complexity/Order of Growth defines the amount of time taken by any program with respect to the size of the input. Time Complexity specifies how the program would behave as the order of size of input is increased. So, Time Complexity is just a function of size of its input. We will go through some of basic and most common time complexities such as:

- *Constant Time Complexity O(1) : constant running time*
- *Linear Time Complexity O(n) : linear running time*
- *Logarithmic Time Complexity O(log n) : logarithmic running time*
- *Log-Linear Time Complexity O(n log n) : log-Â­linear running time*
- *Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)*
- *Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)*

## Constant time Complexity:O(1)
An algorithm is said to have constant time with order O (1) when it is not dependent on the input size n. Irrespective of the input size n, the runtime will always be the same. See example below

```python
# Load the Python libraries
import matplotlib.pyplot as plt
```

```python
# Prob_1
import matplotlib.pyplot as plt

def add_two_numbers(a, b):
    return a + b

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    add_two_numbers(n, n + 1)
    time_values.append(1)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity for Addition")
plt.show()
```

#Linear Time Complexity O(n) : linear running time
- *An algorithm is said to have a linear time complexity when the running time increases linearly with the length of the input. When the function involves checking all the values in input data, with this order O(n).*

```python
def linear_time_sum(numbers):
    total = 0
    for value in numbers:
        total += value
    return total

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    linear_time_sum(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity")
plt.show()
```

#Logarithmic Time Complexity O(log n) : logarithmic running time
- *An algorithm is said to have a logarithmic time complexity when it reduces the size of the input data in each step. This indicates that the number of operations is not the same as the input size. The number of operations gets reduced as the input size increases. Algorithms are found in binary tree or binary search functions. This involves the search of a given value in an array by splitting the array into two and starting searching in one split. This ensures the operation is not done on every element of the data.*

```python
import matplotlib.pyplot as plt   # Import library for plotting graphs
import math                      # Import math library for logarithmic calculations

def binary_search(arr, target):
    """
    Perform binary search on a sorted array to find a target element.
    """
    left, right = 0, len(arr) - 1   # Initialize search boundaries
    steps = 0                       # Counter to track number of steps

    # Continue searching while the left index is less than or equal to right
    while left <= right:
        steps += 1                  # Increment step count each iteration
        mid = (left + right) // 2   # Find the middle index

        # Check if the middle element is the target
        if arr[mid] == target:
            return steps            # Target found, return steps
        # If target is greater, ignore left half
        elif arr[mid] < target:
            left = mid + 1
        # If target is smaller, ignore right half
        else:
            right = mid - 1

    return steps                    # Return steps if target not found

# Generate input sizes from 1 to 1000 with step of 10
n_values = list(range(1, 1000, 10))
time_values = []

# For each input size, approximate time complexity as log2(n)
for n in n_values:
    data = list(range(n))           # Create sorted list of size n
    if n > 0:
        time_values.append(math.log2(n))  # Logarithmic growth
    else:
        time_values.append(0)

# Plotting the logarithmic time complexity curve
plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, label="O(log n)")   # Plot input size vs log2(n)
plt.xlabel("Input size n")                          # Label for x-axis
plt.ylabel("Time (logarithmic scale)")              # Label for y-axis
plt.title("O(log n) Logarithmic Time Complexity (Binary Search)")  # Title
plt.legend()                                        # Show legend
plt.grid(True)                                      # Add grid for clarity
plt.show()                                          # Display the plot
```

#*Log-Linear Time Complexity O(n log n) : log-Â­linear running time*


* When we call a Logarithmic Time Algorithm inside a loop, it would
result into a Log-Linear Time Complexity program.




```python
#Prob_5 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example.
# Prob_5: Log-Linear Time Complexity O(n log n)
import matplotlib.pyplot as plt
import math

# Merge Sort implementation
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = merge_sort(arr[:mid])   # Recursive call on left half
    right = merge_sort(arr[mid:])  # Recursive call on right half

    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0

    # Merge two sorted halves
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    # Add remaining elements
    result.extend(left[i:])
    result.extend(right[j:])
    return result

# Input sizes
n_values = list(range(1, 1000, 50))
time_values = []

# Approximate time complexity as n * log2(n)
for n in n_values:
    if n > 0:
        time_values.append(n * math.log2(n))
    else:
        time_values.append(0)

# Plotting O(n log n)
plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, label="O(n log n)", color="purple")
plt.xlabel("Input size n")
plt.ylabel("Time (log-linear scale)")
plt.title("O(n log n) Log-Linear Time Complexity (Merge Sort Example)")
plt.legend()
plt.grid(True)
plt.show()

```

### Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)

* When the computation time increases as function of N raised to some power, N being the size of input. Such a code has Polynomial Time complexity.




```python
# Prob_6 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example
# Prob_6: Polynomial Time Complexity O(n^2)
import matplotlib.pyplot as plt

def count_pairs(arr):
    """
    Function to count all pairs in the array.
    Time Complexity: O(n^2) because of nested loops.
    """
    count = 0
    n = len(arr)
    for i in range(n):
        for j in range(n):
            count += 1   # Constant work for each pair
    return count

# Input sizes
n_values = list(range(1, 300, 20))   # From 1 to 300 in steps of 20
time_values = []

# Approximate time complexity as n^2
for n in n_values:
    time_values.append(n**2)

# Plotting O(n^2)
plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, label="O(n^2)", color="red")
plt.xlabel("Input size n")
plt.ylabel("Time (polynomial scale)")
plt.title("O(n^2) Polynomial Time Complexity (Nested Loop Example)")
plt.legend()
plt.grid(True)
plt.show()
```

#Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)

* When the computation time of a code increases as function of X^N, N being the size of input. Such a code has Polynomial Time Complexity.




```python
# Prob_7 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example
# Prob_7: Exponential Time Complexity O(2^n)
import matplotlib.pyplot as plt
import math

# Naive recursive Fibonacci function
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

# Input sizes
n_values = list(range(1, 20))   # Keep small, exponential grows fast
time_values = []

# Approximate time complexity as 2^n
for n in n_values:
    time_values.append(2**n)

# Plotting O(2^n)
plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, label="O(2^n)", color="orange")
plt.xlabel("Input size n")
plt.ylabel("Time (exponential scale)")
plt.title("O(2^n) Exponential Time Complexity (Naive Fibonacci Example)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Write a function to perform quick sort and show that it has Logarithmic Time Complexity O(log n)
def quick_sort(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    return quick_sort(left) + middle + quick_sort(right)
```
